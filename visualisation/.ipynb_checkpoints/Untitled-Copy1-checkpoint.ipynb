```python
def autolabel(rects, axes, ajustementY, isPourcentage):
    """ Documentation :
    Afficher les étiquettes de valeur au-dessus de chaque barre. 

    Paramètres : 
        * rects (BarContainer) : objet contenant les rectangles dont on souhaite afficher les valeurs 
        au-dessus
        * axes (matplotlib.axes) : objet permettant de connaître le sous-graphique de la figure où sera 
        afficher le diagramme en barres.
        * ajustementY (int) : en fonction du graphique, la position de l'étiquette au-dessus de chaque barre
        ne sera pas tout à fait la même, donc on fera un ajustement de cette position en donnant en 
        paramètres un entier.  
        * isPourcentage (booléen) : Permet de savoir si on doit afficher les valeurs en pourcentage ou pas 

    Sorties : 
        * rects (BarContainer) : objet contenant les rectangles modifiés
        * axes (matplotlib.axes) : objet permettant de connaître le sous-graphique de la figure où sera 
        afficher le diagramme en barres.
    """
    # Pour chaque barre du diagramme en barres
    for rect in rects:
        # On récupère la hauteur
        height = rect.get_height()
        if (isPourcentage):
            # Affichage de l'étiquette, on arrondit la valeur à un chiffre après la virgule puis on fait
            # quelques réglages de la position et du texte.
            axes.annotate((round(height * 100, 1)), xy=(rect.get_x() + rect.get_width() / 2, height),
                            xytext=(0, ajustementY), textcoords="offset points", ha='center', va='bottom')
        else :
            # On affiche une étiquette que si la valeur n'est pas nulle
            if (height != 0):
                # Affichage de l'étiquette, on arrondit la valeur à un chiffre après la virgule puis on fait
                # quelques réglages de la position et du texte.
                axes.annotate((round(height, 1)), xy=(rect.get_x() + rect.get_width() / 2, height),
                              xytext=(0, ajustementY), textcoords="offset points", ha='center', va='bottom')
    return rects, axes
```

```python
def select_partie1(req):
    res = mycursor.execute(req)
    res = [row for row in res]
    
    nom_elem = [row[0] for row in res]
    nb_elem = [row[1] for row in res]
    
    return (nom_elem, nb_elem)
```

```python
# Initialisation des variables
# cpt = 1
# nbRows = 3
# nbCols = 2
# row = 0

# Création figure
#figure, axes = plt.subplots(nrows=nbRows, ncols=nbCols)

# Taille figure
# figure.set_size_inches(16, 20)

# Titre figure
#figure.suptitle("Display of pie charts and bar charts for qualitative variables",
 #               fontsize=15, color='r')

# Pour chaque attribut, titre et légende
for nb, titre, labels, legende in zip(liste_nb, liste_titre, liste_labels, liste_legende):
    
    # Création figure
    figure, axes = fig, ax = plt.subplots()
    
    

    # Titre ? 
    # 
    
    if (type(labels) == dict):
        labels = [labels[elem] for elem in legende]
        
    # On récupère la proportion de chaque valeur (label) dans la liste 'porpor'
    porpor = np.array(nb) / np.sum(np.array(nb))
     
        
    if (len(labels) > 4):
        # Taille figure
        figure.set_size_inches(15, 10)
        
        # 1) Diagrammes en barres

        # Titre
        axes.set_title("Percentage distribution " +
                              titre, fontsize=13, color='blue')

        # Création diagramme
        rect = axes.bar(labels, porpor, color='darkgray',
                                label='Proportion (%)', edgecolor='black')

        # Ajout des étiquettes au-dessus de chaque barre
        (rect, axes) = autolabel(rect, axes, 2, True)

        # Légende
        axes.legend()

    else :
        # Taille figure
        figure.set_size_inches(10, 10)
        # 2) Diagrammes circulaires

        # Titre
        axes.set_title("Percentage distribution " + titre,
                               fontsize=13, color='blue')

        # Création diagramme 
        wedges, texts, autotexts = axes.pie(
            porpor, labels=labels, radius=1.2,
            autopct=lambda x: str(round(x, 1)) + '%' if x > 1 else '',
            pctdistance=0.7, labeldistance=1.1, shadow=True, 
            colors=['tan', 'indianred', 'mediumslateblue', 'palevioletred',
                    'coral', 'darkseagreen', 'rosybrown', 'darkgray'])

        # Augmentation de la taille du texte
        for text in texts:
            text.set_fontsize('large')

        # Affichage des proportions dans le cercle
        for autotexts in autotexts:
            # En gras
            autotexts.set_weight('bold')
            # Avec un fond rouge
            autotexts.set_bbox(dict(facecolor='white', alpha=0.5))

        # Légende
        axes.legend(wedges, legende, bbox_to_anchor=(1, 0, 0.5, 1))


    # Affichage
    plt.show()
```

```python
def select_partie2(req):
    res = mycursor.execute(req)
    res = [row for row in res]
    
    vehicule = [row[0] for row in res]
    att = [row[1] for row in res]
    count = [row[2] for row in res]
    
    return (vehicule, att, count)
```

```python
def stacked_100(vehicule, att, count, nom_att):
    df = pd.DataFrame(np.transpose(np.array([vehicule, att, count])), columns=['Vehicule', nom_att, 'Value'])
    
    dic = {}
    for vehi in set(vehicule):
        liste = []
        for a in set(att):
            try : 
                nb = int(df.loc[df[nom_att] == a].loc[df['Vehicule'] == vehi].values[0][2])
            except :
                nb = 0 
            liste.append(nb)
        dic[vehi] = liste
    
    df2 = pd.DataFrame.from_dict(dic, orient = 'index', columns=set(att))
    
    df2_100 = df2.div(df2.sum(axis=1), axis=0)
    
    bottom_list = []

    # Loops through each continent column. 
    # Each column will take the sum of all columns before it as its base.

    for i in df2_100.columns:
        b = list(df2_100.loc[:, df2.columns[0]: i].sum(axis=1))
        bottom_list.append(b)
        
    fig, ax = plt.subplots(figsize=(10,10))

    x_ax = df2_100.index

    for i, v in enumerate(df2_100.columns):
        if i == 0:
            # First column is at the bottom of the chart so bottom is 0 and doesn't need to be specified.
            plt.bar(x_ax, df2_100[v])
        else:
            plt.bar(x_ax, df2_100[v], bottom=bottom_list[i - 1])
```
