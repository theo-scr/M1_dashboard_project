```python
def autolabel(rects, axes, ajustementY, isPourcentage):
    """ Documentation :
    Afficher les étiquettes de valeur au-dessus de chaque barre. 

    Paramètres : 
        * rects (BarContainer) : objet contenant les rectangles dont on souhaite afficher les valeurs 
        au-dessus
        * axes (matplotlib.axes) : objet permettant de connaître le sous-graphique de la figure où sera 
        afficher le diagramme en barres.
        * ajustementY (int) : en fonction du graphique, la position de l'étiquette au-dessus de chaque barre
        ne sera pas tout à fait la même, donc on fera un ajustement de cette position en donnant en 
        paramètres un entier.  
        * isPourcentage (booléen) : Permet de savoir si on doit afficher les valeurs en pourcentage ou pas 

    Sorties : 
        * rects (BarContainer) : objet contenant les rectangles modifiés
        * axes (matplotlib.axes) : objet permettant de connaître le sous-graphique de la figure où sera 
        afficher le diagramme en barres.
    """
    # Pour chaque barre du diagramme en barres
    for rect in rects:
        # On récupère la hauteur
        height = rect.get_height()
        if (isPourcentage):
            # Affichage de l'étiquette, on arrondit la valeur à un chiffre après la virgule puis on fait
            # quelques réglages de la position et du texte.
            axes.annotate(str(height) + "%", xy=(rect.get_x() + rect.get_width() / 2, height),
                            xytext=(0, ajustementY), textcoords="offset points", ha='center', va='bottom')
        else :
            # On affiche une étiquette que si la valeur n'est pas nulle
            if (height != 0):
                # Affichage de l'étiquette, on arrondit la valeur à un chiffre après la virgule puis on fait
                # quelques réglages de la position et du texte.
                axes.annotate(str(height), xy=(rect.get_x() + rect.get_width() / 2, height),
                              xytext=(0, ajustementY), textcoords="offset points", ha='center', va='bottom')
    return rects, axes
```

```python
def histo_diag_circu(liste_nb, liste_titre, liste_labels, liste_legende, liste_sauv, isPartie1):
    # Pour chaque attribut, titre et légende
    for nb, titre, labels, legende, sauv in zip(liste_nb, liste_titre, liste_labels, liste_legende, liste_sauv):

        # Création figure
        figure, axes = plt.subplots()



        # Titre ? 
        # 

        if (type(labels) == dict):
            labels = [labels[elem] if elem in labels else elem for elem in legende ]

        # On récupère la proportion de chaque valeur (label) dans la liste 'porpor'
        propor = np.array(nb) / np.sum(np.array(nb))
        propor = [round(item * 100, 1) for item in propor]


        if (len(labels) > 4):
            if (sauv == 'marque') :
                # Taille figure
                figure.set_size_inches(20, 12)
            else : 
                figure.set_size_inches(16, 10)


            # 1) Diagrammes en barres

            # Titre
            axes.set_title("Proportion de citations " +
                                  titre, fontsize=13, color='blue')

            # Création diagramme
            rect = axes.bar(labels, propor, color='darkgray',
                                    label='Proportion (%)', edgecolor='black')

            # Ajout des étiquettes au-dessus de chaque barre
            (rect, axes) = autolabel(rect, axes, 2, True)

            # Légende
            axes.legend()

        else :
            #Taille
            figure.set_size_inches(10, 10)


            # 2) Diagrammes circulaires

            # Titre
            axes.set_title("Proportion de citations " + titre + "\n\n",
                                   fontsize=15, color='blue')

            # Création diagramme 
            wedges, texts, autotexts = axes.pie(
                propor, labels=labels, radius=1.2,
                autopct=lambda x: str(round(x, 1)) + '%' if x > 1 else '',
                pctdistance=0.7, labeldistance=1.1, shadow=True, 
                colors=['tan', 'indianred', 'mediumslateblue', 'palevioletred',
                        'coral', 'darkseagreen', 'rosybrown', 'darkgray'])

            # Augmentation de la taille du texte
            for text in texts:
                text.set_fontsize('large')

            # Affichage des proportions dans le cercle
            for autotexts in autotexts:
                # En gras
                autotexts.set_weight('bold')
                # Avec un fond rouge
                autotexts.set_bbox(dict(facecolor='white', alpha=0.5))

            # Légende
            if (sauv == "moteur"):
                axes.legend(wedges, legende, bbox_to_anchor=(0.6, 0, 0.5, 1))
            else : 
                axes.legend(wedges, legende)


        # Sauvegarde 
        if (isPartie1):
            plt.savefig("partie1_" + sauv + ".jpeg")
        else : 
            plt.savefig("partie4_" + sauv + ".jpeg")
            

        # Affichage
        plt.show()
```

```python
def select_partie2(req):
    res = mycursor.execute(req)
    res = [row for row in res]
    
    vehicule = [row[0] for row in res]
    att = [row[1] for row in res]
    count = [row[2] for row in res]
    
    return (vehicule, att, count)
```

```python
def labels_stacked_100(list_axes):
    
    if len(list_axes) == 2 : 
        for r1, r2 in zip(list_axes[0], list_axes[1]):
            h1 = r1.get_height()
            h2 = r2.get_height()
            if (h1 >0):
                plt.text(r1.get_x() + r1.get_width() / 2., h1 / 2., "%d%%" % h1, ha="center", va="center", color="white", fontsize=16, fontweight="bold")
            if (h2 >0):
                plt.text(r2.get_x() + r2.get_width() / 2., h1 + h2 / 2., "%d%%" % h2, ha="center", va="center", color="white", fontsize=16, fontweight="bold")            
                
    if len(list_axes) == 3 :
        for r1, r2, r3 in zip(list_axes[0], list_axes[1], list_axes[2]):
            h1 = r1.get_height()
            h2 = r2.get_height()
            h3 = r3.get_height()
            if (h1 >0):
                plt.text(r1.get_x() + r1.get_width() / 2., h1 / 2., "%d%%" % h1, ha="center", va="center", color="white", fontsize=16, fontweight="bold")
            if (h2 >0):
                plt.text(r2.get_x() + r2.get_width() / 2., h1 + h2 / 2., "%d%%" % h2, ha="center", va="center", color="white", fontsize=16, fontweight="bold")
            if (h3 >0):
                plt.text(r3.get_x() + r3.get_width() / 2., h1 + h2 + h3 / 2., "%d%%" % h3, ha="center", va="center", color="white", fontsize=16, fontweight="bold")
    
    if len(list_axes) == 4 :
        for r1, r2, r3, r4 in zip(list_axes[0], list_axes[1], list_axes[2], list_axes[3]):
            h1 = r1.get_height()
            h2 = r2.get_height()
            h3 = r3.get_height()
            h4 = r4.get_height()
            if (h1 >0):
                plt.text(r1.get_x() + r1.get_width() / 2., h1 / 2., "%d%%" % h1, ha="center", va="center", color="white", fontsize=16, fontweight="bold")
            if (h2 >0):
                plt.text(r2.get_x() + r2.get_width() / 2., h1 + h2 / 2., "%d%%" % h2, ha="center", va="center", color="white", fontsize=16, fontweight="bold")
            if (h3 >0):
                plt.text(r3.get_x() + r3.get_width() / 2., h1 + h2 + h3 / 2., "%d%%" % h3, ha="center", va="center", color="white", fontsize=16, fontweight="bold")
            if (h4 >0):
                plt.text(r4.get_x() + r4.get_width() / 2., h1 + h2 + h3 + h4 / 2., "%d%%" % h4, ha="center", va="center", color="white", fontsize=16, fontweight="bold")
```

```python
def stacked_100(vehicule, att, count, nom_att, ylabel):
    df = pd.DataFrame(np.transpose(np.array([vehicule, att, count])), columns=['Vehicule', nom_att, 'Value'])
    
    dic = {}
    for vehi in set(vehicule):
        liste = []
        for a in set(att):
            try : 
                nb = int(df.loc[df[nom_att] == a].loc[df['Vehicule'] == vehi].values[0][2])
            except :
                nb = 0 
            liste.append(nb)
        dic[vehi] = liste
    
    df2 = pd.DataFrame.from_dict(dic, orient = 'index', columns=set(att))
    
    df2_100 = df2.div(df2.sum(axis=1), axis=0) * 100
    
    bottom_list = []

    # Loops through each continent column. 
    # Each column will take the sum of all columns before it as its base.

    for i in df2_100.columns:
        b = list(df2_100.loc[:, df2.columns[0]: i].sum(axis=1))
        bottom_list.append(b)
        
    fig, ax = plt.subplots(figsize=(16,10))

    x_ax = df2_100.index
    
    list_axes = []
    
    for enum, c in zip(enumerate(df2_100.columns), colors):
        i = enum[0]
        v = enum[1]
        if i == 0:
            # First column is at the bottom of the chart so bottom is 0 and doesn't need to be specified.
            list_axes.append(ax.bar(x_ax, df2_100[v], label=v, color=c, width=0.65))
        else:
            list_axes.append(ax.bar(x_ax, df2_100[v], bottom=bottom_list[i - 1], label=v, color=c, width=0.65))
            
    labels_stacked_100(list_axes)

    
    plt.xlabel("\nTypes de véhicule", fontsize=13)
    plt.ylabel("Répartition (en %) "+ ylabel, fontsize=13)
    
    plt.legend()
    
    # Sauvegarde 
    plt.savefig("partie2_" + nom_att + ".jpeg")
```

```python
def timeline(title, list_colors, annees, atts):

    py.init_notebook_mode(connected=True)

    df = []
    for annee, att in zip(annees, atts):
        dic = dict(Task=att, Start=annee, Finish=str(int(annee)+1), Resource=att)
        df.append(dic)
        
    colors = {}
    for color, resource in zip(list_colors, set(atts)):
        colors[resource] = color
        
    
    if len(colors) > 5:
        height = 700
    else : 
        height = 500
    
    fig = ff.create_gantt(df, colors = colors, index_col='Resource', show_colorbar=True, group_tasks=True,
                          bar_width=0.2, showgrid_x=True, showgrid_y=True, title  =title, 
                         show_hover_fill =False, height=height)
    
    py.iplot(fig, filename='stacked-bar')
```

```python
def autolabel2(rects, labels, axes, ajustementY, isPourcentage):
    """ Documentation :
    Afficher les étiquettes de valeur au-dessus de chaque barre. 

    Paramètres : 
        * rects (BarContainer) : objet contenant les rectangles dont on souhaite afficher les valeurs 
        au-dessus
        * axes (matplotlib.axes) : objet permettant de connaître le sous-graphique de la figure où sera 
        afficher le diagramme en barres.
        * ajustementY (int) : en fonction du graphique, la position de l'étiquette au-dessus de chaque barre
        ne sera pas tout à fait la même, donc on fera un ajustement de cette position en donnant en 
        paramètres un entier.  
        * isPourcentage (booléen) : Permet de savoir si on doit afficher les valeurs en pourcentage ou pas 

    Sorties : 
        * rects (BarContainer) : objet contenant les rectangles modifiés
        * axes (matplotlib.axes) : objet permettant de connaître le sous-graphique de la figure où sera 
        afficher le diagramme en barres.
    """
    # Pour chaque barre du diagramme en barres
    for rect, label in zip(rects, labels):
        # On récupère la hauteur
        height = rect.get_height()
        if (isPourcentage):
            # Affichage de l'étiquette, on arrondit la valeur à un chiffre après la virgule puis on fait
            # quelques réglages de la position et du texte.
            axes.annotate(str(label), xy=(rect.get_x() + rect.get_width() / 2, height),
                          xytext=(0, ajustementY), textcoords="offset points", ha='center',
                          va='bottom', fontsize=15)
    return rects, axes
```

```python
#, COUNT(A2.id_article)
req_sql = " SELECT P2.nom_pays, V2.type_vehicule \
            FROM Article A2, Vehicule V2, Vehicule_Evoq VE2, Pays P2, Pays_Evoq PE2 \
            WHERE A2.id_article = VE2.id_article AND V2.id_vehicule = VE2.id_vehicule AND A2.id_article = PE2.id_article AND P2.id_pays = PE2.id_pays  \
            AND P2.nom_pays IN ( \
                SELECT * FROM ( \
                    SELECT P.nom_pays \
                    FROM Pays_Evoq PE, Pays P \
                    WHERE P.id_pays = PE.id_pays \
                    GROUP BY P.nom_pays \
                    ORDER BY COUNT(PE.id_pays) DESC \
                ) \
                WHERE ROWNUM <= 10) \
            AND V2.type_vehicule IN ( \
                SELECT * FROM ( \
                    SELECT V.type_vehicule \
                    FROM Vehicule V, Vehicule_Evoq VE, Article A, Pays_Evoq PE \
                    WHERE V.id_vehicule = VE.id_vehicule AND A.id_article = PE.id_article AND A.id_article = VE.id_article AND PE.id_pays = PE2.id_pays \
                    GROUP BY V.type_vehicule \
                    ORDER BY COUNT(VE.id_vehicule) DESC \
                ) \
                WHERE ROWNUM <= 3) \
            GROUP BY P2.nom_pays, V2.type_vehicule \
            ORDER BY P2.nom_pays, COUNT(A2.id_article) DESC"

(pays, vehicules) = select_partie1(req_sql)
```

```python
def podium(pays, vehicules, sauv, titre, xlabel, ylabel):
    
    #plt.figure(figsize=(20,10))
    fig, ax = plt.subplots(figsize=(20,10))
    
    y1 = [2] * len(set(pays))
    y2 = [3] * len(set(pays))
    y3 = [1] * len(set(pays))
    
    barWidth = 0.3
    
    r1 = range(len(y1))
    r2 = [x + barWidth for x in r1]
    r3 = [x + barWidth for x in r2]
    
    rects1 = ax.bar(r1, y1, width = barWidth, color = ['silver' for i in y1], linewidth = 2)
    rects2 = ax.bar(r2, y2, width = barWidth, color = ['gold' for i in y1], linewidth = 4)
    rects3 = ax.bar(r3, y3, width = barWidth, color = ['tan' for i in y1], linewidth = 4)
    
    labels1 = [item for item in vehicules[::3]]
    labels2 = [item for item in vehicules[1::3]]
    labels3 = [item for item in vehicules[2::3]]
    
    labelsx = [item for item in pays[::3]]
    
    plt.xticks([r + 2*barWidth / 3 for r in range(len(y1))], labelsx, fontsize=14)
    
    
    (rect, ax) = autolabel2(rects1, labels2, ax, 2, True)
    (rect, ax) = autolabel2(rects2, labels1, ax, 2, True)
    (rect, ax) = autolabel2(rects3, labels3, ax, 2, True)
    
    plt.title(titre, fontsize=15)
    
    plt.xlabel(xlabel, fontsize=15)
    plt.ylabel(ylabel, fontsize=15)
    
    
    plt.savefig("partie4_" + sauv + ".jpeg")
    
    plt.show()
```
